### **DATA SPLITING AND SCALING**

```python
X=df.iloc[:,0:7].values
Y=df.iloc[:,-1].values
```

```python
# Train test split
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=101)
```

```python
# Standard scaling our data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train) 
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape , X_test.shape
```

## **MODEL BUILDING** 

#### **1. LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score
from sklearn import preprocessing
lg = LogisticRegression(max_iter=1000)
lg.fit(X_train, Y_train)
```

```python
Y_pred = lg.predict(X_test)
c1= lg.score(X_test, Y_test)
p1 = precision_score(Y_test, Y_pred)
```

```python
cf_matrix = confusion_matrix(Y_test, Y_pred)
sns.heatmap(cf_matrix, annot=True,cmap='Blues')
plt.show()
```

#### **2. DECISION TREE CLASSIFIER**

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, Y_train)
```

```python
Y_pred = dtc.predict(X_test)
c2= dtc.score(X_test, Y_test)
p2 = precision_score(Y_test, Y_pred)
```

```python
cf_matrix = confusion_matrix(Y_test, Y_pred)
sns.heatmap(cf_matrix, annot=True,cmap='Blues')
plt.show()
```

#### **3. RANDOM FOREST CLASSIFIER**

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, Y_train)
```

```python
Y_pred = rfc.predict(X_test)
c3= rfc.score(X_test, Y_test)
p3 = precision_score(Y_test, Y_pred)
```

```python
cf_matrix = confusion_matrix(Y_test, Y_pred)
sns.heatmap(cf_matrix, annot=True,cmap='Blues')
plt.show()
```

#### **4. K-NEAREST NEIGHBOR**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7) 
knn.fit(X_train, Y_train)
```

```python
Y_pred = knn.predict(X_test)
c4= knn.score(X_test, Y_test)
p4 = precision_score(Y_test, Y_pred)
```

```python
cf_matrix = confusion_matrix(Y_test, Y_pred)
sns.heatmap(cf_matrix, annot=True,cmap='Blues')
plt.show()
```

#### **5. SVC**

```python
from sklearn import svm
sv = svm.SVC()
sv.fit(X_train, Y_train)
```

```python
Y_pred = sv.predict(X_test)
c5= sv.score(X_test, Y_test)
p5 = precision_score(Y_test, Y_pred)
```

```python
cf_matrix = confusion_matrix(Y_test, Y_pred)
sns.heatmap(cf_matrix, annot=True,cmap='Blues')
plt.show()
```

#### **6. GAUSSIAN NAIVE BAYES**

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, Y_train)
```

```python
Y_pred = gnb.predict(X_test)
c6= gnb.score(X_test, Y_test)
p6 = precision_score(Y_test, Y_pred)
```

```python
cf_matrix = confusion_matrix(Y_test, Y_pred)
sns.heatmap(cf_matrix, annot=True,cmap='Blues')
plt.show()
```

## **MODEL COMPARISION**

```python
name = ('LOGISTIC REGRESSION', 'DECISION TREE CLASSIFIER','RANDOM FOREST CLASSIFIER','K-NEAREST NEIGHBOR','SVC','GAUSSIAN NAIVE BAYES')
accuracy_score = (c1,c2,c3,c4,c5,c6)
precision_score = (p1,p2,p3,p4,p5,p6)
```

```python
dfd = pd.DataFrame()
dfd['Name']= name
dfd['Accuracy Score']= accuracy_score
dfd['Precision Score']= precision_score
dfd
```

```python
sns.barplot(x = "Name", y = 'Accuracy Score', data = dfd, palette = 'viridis')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(x = "Name", y = 'Precision Score', data = dfd, palette = 'viridis')
plt.xticks(rotation=90)
plt.show()
```
